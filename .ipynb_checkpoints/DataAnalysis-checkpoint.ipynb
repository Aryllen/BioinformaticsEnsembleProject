# Data Analysis

This notebok analyses the data in the super matrices created in the Data Wrangling notebook.

In this notebook we find which genes were selected most often by the classifiers, create histograms showing the frequency distribution, and compare genes using a spearman rank correlation.

```r
# Read in files
coef.super.matrix.10k = read.table("super_matrix/coef_10k.txt", sep="\t", header=T)
coef.super.matrix.5k = read.table("super_matrix/coef_5k.txt", sep="\t", header=T)
coef.super.matrix.2.5k = read.table("super_matrix/coef_2.5k.txt", sep="\t", header=T)
coef.super.matrix.1k = read.table("super_matrix/coef_1k.txt", sep="\t", header=T)
coef.super.matrix.100 = read.table("super_matrix/coef_100.txt", sep="\t", header=T)
```

```r
# sanity check
dim(coef.super.matrix.10k)
dim(coef.super.matrix.5k)
dim(coef.super.matrix.2.5k)
dim(coef.super.matrix.1k)
dim(coef.super.matrix.100)

coef.super.matrix.10k[, 1:10]
coef.super.matrix.5k[, 1:10]
coef.super.matrix.2.5k[, 1:10]
coef.super.matrix.1k[, 1:10]
coef.super.matrix.100[, 1:10]
```

### Which genes were picked most often by classifiers?

```r
# Sum up the times a gene was chosen (i.e. coefficient/importance != 0)
sums.10k <- apply(coef.super.matrix.10k, 2, function(x){
    ml.sum <- 0
    for (i in x) {
        if (i != 0) {
            ml.sum <- ml.sum + 1
        }
    }
    return (ml.sum)
})

sums.5k <- apply(coef.super.matrix.5k, 2, function(x){
    ml.sum <- 0
    for (i in x) {
        if (i != 0) {
            ml.sum <- ml.sum + 1
        }
    }
    return (ml.sum)
})

sums.2.5k <- apply(coef.super.matrix.2.5k, 2, function(x){
    ml.sum <- 0
    for (i in x) {
        if (i != 0) {
            ml.sum <- ml.sum + 1
        }
    }
    return (ml.sum)
})

sums.1k <- apply(coef.super.matrix.1k, 2, function(x){
    ml.sum <- 0
    for (i in x) {
        if (i != 0) {
            ml.sum <- ml.sum + 1
        }
    }
    return (ml.sum)
})

sums.100 <- apply(coef.super.matrix.100, 2, function(x){
    ml.sum <- 0
    for (i in x) {
        if (i != 0) {
            ml.sum <- ml.sum + 1
        }
    }
    return (ml.sum)
})
```

```r
# Sanity check. First in sums should be the same number as ML that chose the gene
sums.10k[1:5]
coef.super.matrix.10k[,1:2]
```

```r
# Find the indices where more than one model chose the gene
sums.not.zero.10k <- sums.10k[which(sums.10k != 0)]
sums.not.zero.5k <- sums.5k[which(sums.5k != 0)]
sums.not.zero.2.5k <- sums.2.5k[which(sums.2.5k != 0)]
sums.not.zero.1k <- sums.1k[which(sums.1k != 0)]
sums.not.zero.100 <- sums.100[which(sums.100 != 0)]
```

```r
# Print histograms
par(mfrow=c(3,2))
hist(sums.not.zero.10k, main="10k Genes", xlab="Number of ML Algorithms")
hist(sums.not.zero.5k, main="5k Genes", xlab="Number of ML Algorithms")
hist(sums.not.zero.2.5k, main="2.5k Genes", xlab="Number of ML Algorithms")
hist(sums.not.zero.1k, main="1k Genes", xlab="Number of ML Algorithms")
hist(sums.not.zero.100, main="100 Genes", xlab="Number of ML Algorithms")
dev.print(png, file="Hist_MLAlgs_per_Gene_g0.png", width = 500, height = 500, units = "px")
```

### Spearman Rank Correlation

##### Compare gene expression for genes selected 15 or more times by models

```r
# Load original, normalized data in t-test reduced set; using 10k dataset since all other datasets should be included
# RNA data
data.rna <- read.table("aml.data.RNA.reduced.csv", sep=",", header=TRUE)
# Make entity_id the row names of the table
dimnames(data.rna)[[1]] <- data.rna[,1]
data.rna <- data.rna[,-1]
data.rna[1:5,1:5]
# Patient labels
# Do not make entity_id the row names or it will drop it to a vector and remove the names
data.labels <- read.table("aml.data.labels.csv", sep=",", header=TRUE)
data.labels[1:5,]
```

```r
# Dataframe to hold results
rank.cor.results.10k <- data.frame(matrix(nrow=length(indices.10k),ncol=3))
dimnames(rank.cor.results.10k)[[2]] <- c("statistic", "p.value", "estimate")
dimnames(rank.cor.results.10k)[[1]] <- dimnames(data.rna)[[2]][indices.10k]
rank.cor.results.10k[1:5,]

# Dataframe to hold results
rank.cor.results.5k <- data.frame(matrix(nrow=length(indices.5k),ncol=3))
dimnames(rank.cor.results.5k)[[2]] <- c("statistic", "p.value", "estimate")
dimnames(rank.cor.results.5k)[[1]] <- dimnames(data.rna)[[2]][indices.5k]
rank.cor.results.5k[1:5,]

# Dataframe to hold results
rank.cor.results.2.5k <- data.frame(matrix(nrow=length(indices.2.5k),ncol=3))
dimnames(rank.cor.results.2.5k)[[2]] <- c("statistic", "p.value", "estimate")
dimnames(rank.cor.results.2.5k)[[1]] <- dimnames(data.rna)[[2]][indices.2.5k]
rank.cor.results.2.5k[1:5,]

# Dataframe to hold results
rank.cor.results.1k <- data.frame(matrix(nrow=length(indices.1k),ncol=3))
dimnames(rank.cor.results.1k)[[2]] <- c("statistic", "p.value", "estimate")
dimnames(rank.cor.results.1k)[[1]] <- dimnames(data.rna)[[2]][indices.1k]
rank.cor.results.1k[1:5,]

# Dataframe to hold results
rank.cor.results.100 <- data.frame(matrix(nrow=length(indices.100),ncol=3))
dimnames(rank.cor.results.100)[[2]] <- c("statistic", "p.value", "estimate")
dimnames(rank.cor.results.100)[[1]] <- dimnames(data.rna)[[2]][indices.100]
rank.cor.results.100[1:5,]
```

```r
# Compare gene expression for genes selected most by models
for (i in indices.10k) {
    rank.cor.temp <- cor.test(data.rna[,i], data.labels[,2], method="spearman", exact=FALSE)
    gene.id <- dimnames(data.rna)[[2]][i]
    gene.index <- which(dimnames(rank.cor.results.10k)[[1]] == gene.id)
    rank.cor.results.10k[gene.index, 1] <- rank.cor.temp$statistic
    rank.cor.results.10k[gene.index, 2] <- rank.cor.temp$p.value
    rank.cor.results.10k[gene.index, 3] <- rank.cor.temp$estimate
}

# Compare gene expression for genes selected most by models
for (i in indices.5k) {
    rank.cor.temp <- cor.test(data.rna[,i], data.labels[,2], method="spearman", exact=FALSE)
    gene.id <- dimnames(data.rna)[[2]][i]
    gene.index <- which(dimnames(rank.cor.results.5k)[[1]] == gene.id)
    rank.cor.results.5k[gene.index, 1] <- rank.cor.temp$statistic
    rank.cor.results.5k[gene.index, 2] <- rank.cor.temp$p.value
    rank.cor.results.5k[gene.index, 3] <- rank.cor.temp$estimate
}

# Compare gene expression for genes selected most by models
for (i in indices.2.5k) {
    rank.cor.temp <- cor.test(data.rna[,i], data.labels[,2], method="spearman", exact=FALSE)
    gene.id <- dimnames(data.rna)[[2]][i]
    gene.index <- which(dimnames(rank.cor.results.2.5k)[[1]] == gene.id)
    rank.cor.results.2.5k[gene.index, 1] <- rank.cor.temp$statistic
    rank.cor.results.2.5k[gene.index, 2] <- rank.cor.temp$p.value
    rank.cor.results.2.5k[gene.index, 3] <- rank.cor.temp$estimate
}

# Compare gene expression for genes selected most by models
for (i in indices.1k) {
    rank.cor.temp <- cor.test(data.rna[,i], data.labels[,2], method="spearman", exact=FALSE)
    gene.id <- dimnames(data.rna)[[2]][i]
    gene.index <- which(dimnames(rank.cor.results.1k)[[1]] == gene.id)
    rank.cor.results.1k[gene.index, 1] <- rank.cor.temp$statistic
    rank.cor.results.1k[gene.index, 2] <- rank.cor.temp$p.value
    rank.cor.results.1k[gene.index, 3] <- rank.cor.temp$estimate
}

# Compare gene expression for genes selected most by models
for (i in indices.100) {
    rank.cor.temp <- cor.test(data.rna[,i], data.labels[,2], method="spearman", exact=FALSE)
    gene.id <- dimnames(data.rna)[[2]][i]
    gene.index <- which(dimnames(rank.cor.results.100)[[1]] == gene.id)
    rank.cor.results.100[gene.index, 1] <- rank.cor.temp$statistic
    rank.cor.results.100[gene.index, 2] <- rank.cor.temp$p.value
    rank.cor.results.100[gene.index, 3] <- rank.cor.temp$estimate
}
```

```r
# Sanity Check
dim(rank.cor.results.10k)
rank.cor.results.10k[1:5,]
dim(rank.cor.results.5k)
rank.cor.results.5k[1:5,]
dim(rank.cor.results.2.5k)
rank.cor.results.2.5k[1:5,]
dim(rank.cor.results.1k)
rank.cor.results.1k[1:5,]
dim(rank.cor.results.100)
rank.cor.results.100[1:5,]
```

### Accuracy averages (Alex's code below)

```r
# coef.super.matrix.10k = read.table("super_matrix/coef_10k.txt", sep="\t", header=T)
acc.10k = read.table("super_matrix/acc_10k.txt", sep="\t", header=T)
acc.5k = read.table("super_matrix/acc_5k.txt", sep="\t", header=T)
acc.2.5k = read.table("super_matrix/acc_2.5k.txt", sep="\t", header=T)
acc.1k = read.table("super_matrix/acc_1k.txt", sep="\t", header=T)
acc.100 = read.table("super_matrix/acc_100.txt", sep="\t", header=T)

avg.10k = vector(mode="numeric", length=ncol(acc.10k))
avg.5k = vector(mode="numeric", length=ncol(acc.5k))
avg.2.5k = vector(mode="numeric", length=ncol(acc.2.5k))
avg.1k = vector(mode="numeric", length=ncol(acc.1k))
avg.100 = vector(mode="numeric", length=ncol(acc.100))

for (i in 1:length(avg.10k)) {
    avg.10k[i] = mean(acc.10k[,i])
}
for (i in 1:length(avg.5k)) {
    avg.5k[i] = mean(acc.5k[,i])
}
for (i in 1:length(avg.2.5k)) {
    avg.2.5k[i] = mean(acc.2.5k[,i])
}
for (i in 1:length(avg.1k)) {
    avg.1k[i] = mean(acc.1k[,i])
}
for (i in 1:length(avg.100)) {
    avg.100[i] = mean(acc.100[,i])
}
avg.10k
avg.5k
avg.2.5k
avg.1k
avg.100
```
